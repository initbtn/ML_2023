```python
import torch
import torch.optim as optim
```

```python
torch.manual_seed(1)
```

```python
x_data=torch.FloatTensor([[1],[2],[3],[4],[5]])
t_data=torch.FloatTensor([[3],[5],[7],[9],[11]])
```

```python
w=torch.zeros(1, requires_grad=True)
b=torch.zeros(1, requires_grad=True)
print(w, b)
```

```python
y=x_data*w+b
```

```python
cost=torch.mean((t_data-y)**2)
print(cost)
```

```python
optimizer=optim.SGD([w,b], lr=0.01) # 경사하강법으로 w,b 업데이트하기위한 최적화 식
```

```python
x_data=torch.FloatTensor([[1],[2],[3],[4],[5]])
t_data=torch.FloatTensor([[3],[5],[7],[9],[11]])

w=torch.zeros(1, requires_grad=True)
b=torch.zeros(1, requires_grad=True)

optimizer=optim.SGD([w,b], lr=0.01)

nb_epochs=2000
for epoch in range(nb_epochs+1):
    y=x_data*w+b
    cost=torch.mean((y-t_data)**2)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch%100==0:
        print('Epoch {:4d} w: {:.3f}, b:{:.3f} Cost:{:.6f}'.format(
            epoch, w.item(), b.item(), cost.item()))
        
```

# multi liner regression

```python
x_data=torch.FloatTensor([[73,  80,  75], 
                              [93,  88,  93], 
                               [89,  91,  80], 
                               [96,  98,  100],   
                               [73,  66,  70]]) 
t_data=y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])

W=torch.zeros((3, 1), requires_grad=True)
b=torch.zeros(1, requires_grad=True)

optimizer=optim.SGD([W,b], lr=1e-6)
nb_epochs=2000
for epoch in range(nb_epochs+1):
    y=x_data.matmul(W)+b
    cost=torch.mean((y-t_data)**2)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    print('Epoch: {:4d}/{} Hypothesis:{} Cost:{:.6f}'.format(
          epoch, nb_epochs, y.squeeze().detach(),cost.item()))
```
