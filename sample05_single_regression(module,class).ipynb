```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
```

```python
torch.manual_seed(1)
x_data=torch.FloatTensor([[1],[2],[3],[4],[5]])
t_data=torch.FloatTensor([[3],[5],[7],[9],[11]])
```

```python
model=nn.Linear(1,1) # input_dim=1, output_dim=1
print(list(model.parameters()))
```

```python
optimizer=optim.SGD(model.parameters(), lr=0.01)
```

```python
nb_epochs=20000

for epoch in range(nb_epochs):
    prediction=model(x_data)
    cost=F.mse_loss(prediction, t_data)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch%400==0:
        print('Epoch:', epoch, 'Cost:',cost.item())
```

```python
new_data=torch.FloatTensor([[10]])
y_pred=model(new_data)
print(y_pred)
print(list(model.parameters()))
```

# class 사용 single regression

```python
torch.manual_seed(1)
x_data=torch.FloatTensor([[1],[2],[3],[4],[5]])
t_data=torch.FloatTensor([[3],[5],[7],[9],[11]])
```

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear=nn.Linear(1,1)

    def forward(self, x):
        return self.linear(x)
```

```python
model=LinearRegressionModel()
print(list(model.parameters()))
optimizer=optim.SGD(model.parameters(), lr=0.01)
```

```python
nb_epochs=20000
for epoch in range(nb_epochs):
    prediction=model(x_data)
    cost=F.mse_loss(prediction, t_data)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch%400==0:
        print('Epoch', epoch, 'Cost:', cost.item())
```

```python
new_data=torch.FloatTensor([[10]])
pred_y=model(new_data)
print(pred_y)
print(list(model.parameters()))
```
